# What are the most used words

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern , book.lower())                   
```

```python
findings[0:5]
```

```python
Dict ={}
```

```python
for word in findings:
    if word in Dict.keys():
        Dict[word] =Dict[word]+1
    else:
        Dict[word] =1
```

# The most used words(non_articles)

```python
Dict_list =[(value,key) for (key,value) in Dict.items()]
```

```python
import nltk
```

```python
Dict_list = sorted(Dict_list , reverse= True)
```

```python
from nltk.corpus import stopwords
english_words = stopwords.words("english")
filtered_words =[]
for count,word in Dict_list:
    if word  not in english_words:
        filtered_words.append((word,count))
        
```

```python
filtered_words[0:5]
```

# Sentiment Analysis, What is most postive and negative Chapter ?


```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern,book)
chapters = chapters[1:]
```

```python

for i,chapter in enumerate(chapters ,start =1):
    score = analyzer.polarity_scores(chapter)
    if score["pos"] > score["neg"]:
        print(f"Chapter {i} is Positive Chapter.")
        print(score)
    else:
        print(f"Chapter {i} is Negative Chapter.") 
        print(score)

    
```
